```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
x = np.array(([1.303384, 0], [2.304527, 0], [3.30569, 9], [4.306939, 9], [5.308143, 13], [6.309362, 13], [7.309938, 16]))

time = [data[0] for data in x]    
quality = [data[1] for data in x]    

plt.plot(time, quality, marker='o')

# Adding titles and labels
plt.title('Quality vs. Time')
plt.xlabel('Time')
plt.ylabel('Quality')
```
